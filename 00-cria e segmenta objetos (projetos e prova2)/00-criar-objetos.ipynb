```python
# fonte: https://richardpricejones.medium.com/drawing-a-rectangle-with-a-angle-using-opencv-c9284eae3380
def draw_angled_rec(x0, y0, width, height, angle, cor, tam, img):
  import math
  _angle = angle * math.pi / 180.0
  b = math.cos(_angle) * 0.5
  a = math.sin(_angle) * 0.5
  pt0 = (int(x0 - a * height - b * width),
          int(y0 + b * height - a * width))
  pt1 = (int(x0 + a * height - b * width),
          int(y0 - b * height - a * width))
  pt2 = (int(2 * x0 - pt0[0]), int(2 * y0 - pt0[1]))
  pt3 = (int(2 * x0 - pt1[0]), int(2 * y0 - pt1[1]))

  cv2.line(img, pt0, pt1, cor, tam)
  cv2.line(img, pt1, pt2, cor, tam)
  cv2.line(img, pt2, pt3, cor, tam)
  cv2.line(img, pt3, pt0, cor, tam)
  return img
```

# 2 - Objetos aleatórios

Preparando uma base de dados para ser utilizada no treinamento de Redes Neurais Convolucionais (CNN), no reconhecimento de objetos

```python
import numpy as np
import cv2

#np.random.seed(42)

def createRandomImages(N=10, num_objects_type=3, ruido=0.78, tipo='test'):

  # Definir altura e largura da imagem
  width = 608
  height = 608

  # Definir distância mínima entre objetos
  min_distance = 80

  size_min = 20
  size_max = min_distance//2

  # Definir espessura da borda
  border_thickness = 2
  for cont in range(N): # cria N imagens
    # Criar uma lista para armazenar as coordenadas dos centros dos objetos
    centers = []
    border = 60
    i=0

    rotulos = ""

    # Criar imagem branca
    img = np.zeros((height, width, 3), dtype=np.uint8)

    # Criar imagem label para cada objeto branca
    imgTriangulo = np.zeros((height, width), dtype=np.uint8)
    imgQuadrado = imgTriangulo.copy()
    imgPendagono = imgTriangulo.copy()
    imgHexagono = imgTriangulo.copy()
    imgHeptagono = imgTriangulo.copy()
    imgCirculo = imgTriangulo.copy()
    imgElipse = imgTriangulo.copy()

    num_objetos_desenhados = np.zeros(len(objetos), dtype=np.uint8)
    for obj_type in objetos:
      imgTipo = np.zeros((height, width), dtype=np.uint8)

      print(obj_type)
      while num_objetos_desenhados[objetos.index(obj_type)] < num_objects_type:
        #print(*num_objetos_desenhados)

        # Definir um centro aleatório para o objeto
        center = (np.random.randint(border,width-border), np.random.randint(border,height-border))

        # Verificar se o objeto está muito próximo de outros objetos
        too_close = False
        for c in centers:
          if np.sqrt((center[0] - c[0]) ** 2 + (center[1] - c[1]) ** 2) < min_distance:
            too_close = True
            break
        if too_close:
          continue

        cor = np.random.randint(50, 256, size=(3)).tolist()

        # Adicionar o centro à lista de centros
        centers.append(center)

        # Definir um ângulo aleatório para a rotação do objeto
        angle = np.random.randint(360)

        # Desenhar o objeto
        if obj_type in objetos[:-2]: # 'polygon'
          # Definir um número de lados para o polígono
          sides = objetos.index(obj_type)+3

          # Definir um tamanho aleatório para o polígono
          size = np.random.randint(size_min, size_max)

          # Definir as coordenadas dos vértices do polígono
          vertices = []
          for j in range(sides):
            theta = (2 * np.pi / sides) * j
            x = center[0] + size * np.cos(theta + np.radians(angle))
            y = center[1] + size * np.sin(theta + np.radians(angle))
            vertices.append((int(x), int(y)))

          # desenha image e labels
          if num_objetos_desenhados[objetos.index(obj_type)] < num_objects_type:
            cv2.fillPoly(img, [np.array(vertices)], cor, lineType=cv2.LINE_AA)
            cv2.fillPoly(eval('img'+objetos[objetos.index(obj_type)]), [np.array(vertices)], 255, lineType=cv2.LINE_AA)
            num_objetos_desenhados[objetos.index(obj_type)] += 1

          x,y = center
          rotulos += f"{objetos.index(obj_type)}\t{(x-size)/width:.4f}\t{(y-size)/height:.4f}\t{(x+size)/width:.4f}\t{(y+size)/height:.4f}\n"
          # para testar
          #cv2.rectangle(img,(int(x-size),int(y-size)),(int(x+size),int(y+size)),(255,255,255),2)

        else: # 'ellipse'
          if obj_type == objetos[-2]: # círculo
            # Definir os eixos maiores e menores do "círculo"
            while True:
              major_axis = np.random.randint(size_min, size_max)
              minor_axis = np.random.randint(size_min, size_max)
              if abs(major_axis-minor_axis)<5:
                break

            # desenha image e labels
            cv2.ellipse(img, center, (major_axis, minor_axis), angle, 0, 360, cor, -1, lineType=cv2.LINE_AA)
            cv2.ellipse(imgCirculo, center, (major_axis, minor_axis), angle, 0, 360, 255, -1, lineType=cv2.LINE_AA)
            num_objetos_desenhados[objetos.index(obj_type)] += 1
          elif obj_type == objetos[-1]: # elipse
            # Definir os eixos maiores e menores da elipse
            while True:
              major_axis = np.random.randint(size_min, size_max)
              minor_axis = np.random.randint(size_min, size_max)
              if abs(major_axis-minor_axis)>5:
                break
            cv2.ellipse(img, center, (major_axis, minor_axis), angle, 0, 360, cor, -1, lineType=cv2.LINE_AA)
            cv2.ellipse(imgElipse, center, (major_axis, minor_axis), angle, 0, 360, 255, -1, lineType=cv2.LINE_AA)
            num_objetos_desenhados[objetos.index(obj_type)] += 1

          x,y = center
          m = max(minor_axis,major_axis)
          rotulos += f"{objetos.index(obj_type)}\t{(x-m)/width:.4f}\t{(y-m)/height:.4f}\t{(x+m)/width:.4f}\t{(y+m)/height:.4f}\n"
          # para testar
          #cv2.rectangle(img,(int(x-m),int(y-m)),(int(x+m),int(y+m)),(255,255,255),2)

    # ajustar os ruídos
    row, col, ch = img.shape
    mean = 0
    var = 500
    sigma = var**ruido
    noise = np.random.normal(mean, sigma, (row, col, ch))
    img = np.clip(img + noise, 0, 255)

    fname = str(cont).zfill(5)
    _=cv2.imwrite(f"data/{tipo}/image/{fname}.jpg",img)
    with open(f"data/{tipo}/image/{fname}.txt", 'w') as f:
      f.write(rotulos)

    for o in objetos:
      _=cv2.imwrite(f"data/{tipo}/{o}/{fname}.jpg",eval("img" + o))
```

```python
# classes: 0, 1, 2, 3, 4, 5 e 6
objetos = ['Triangulo','Quadrado','Pendagono','Hexagono','Heptagono','Circulo','Elipse']
```

```python
# criar pastas
import os
from os import path
pastas = ['output',
          'data',
          'data/train',
          'data/train/image',
          'data/test',
          'data/test/image']
for o in objetos:
  pastas.append('data/train/'+o)
  pastas.append('data/test/'+o)
pastas = sorted(pastas)
for p in pastas:
  if not path.exists(p):
    os.mkdir(p)
  print(p)
```
