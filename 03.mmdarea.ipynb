```python
img = mm.read('https://drive.google.com/file/d/1z1oThl9hPxZV-5c7u6HlDEzqEAaRRFqm/view?usp=sharing')
f = cv2.cvtColor(img ,cv2.COLOR_BGR2GRAY) # converte para grayscale
bin = np.array(f>30).astype('uint8')
plt.figure(figsize=(10, 10))
mm.show(bin)  
```

```python
areaopen = mm.areaopen(bin,180)
plt.figure(figsize=(10, 10))
mm.show(areaopen>0) 
```

```python
plt.figure(figsize=(10, 10))
mm.show(img,areaopen) 
```

```python
import numpy as np
img2 = mm.open(bin,np.ones((20,1),dtype='uint8'))
mm.show(img,img2)
```

```python
img3 = mm.infrec(img2,bin)
mm.show(img,img3)
```
