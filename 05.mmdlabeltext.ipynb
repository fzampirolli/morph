# mmdlabeltext

```python
img = mm.read('https://drive.google.com/file/d/1v4JoI_vF5HcCBn6ts2YJgOs0T7qDWUGF/view?usp=sharing')
f = cv2.cvtColor(img ,cv2.COLOR_BGR2GRAY) # converte para grayscale
f = mm.threshold(f)
plt.figure(figsize=(10, 10))
mm.show(f)  
```

```python
f1 = mm.close(f,np.ones((3,1), dtype='uint8'))
f1 = np.array(mm.areaopen(f1,10)>0, dtype='uint8')
plt.figure(figsize=(10, 10))
mm.show(f1)  
```

```python
labels_letra = mm.label(f1)
plt.figure(figsize=(10, 10))
mm.show(labels_letra)  
np.amax(labels_letra)
```

```python
b = np.ones((3,7),dtype='uint8')

open = mm.close(f,b)
labels_palavra = mm.label(open)

plt.figure(figsize=(10, 10))
mm.show(labels_palavra)  
```

```python
b = np.ones((9,19),dtype='uint8')

open = mm.close(f,b)
labels_paragrafo = mm.label(open)

plt.figure(figsize=(10, 10))
mm.show(labels_paragrafo)  
```

```python
parag = np.minimum(f,labels_paragrafo)

plt.figure(figsize=(10, 10))
mm.show(parag)
```
