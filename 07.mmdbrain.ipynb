```python
img = mm.read('https://drive.google.com/file/d/1HpmTEryzshUWdEkMZt9ac824BYgSqzPr/view?usp=sharing')
a = cv2.cvtColor(img ,cv2.COLOR_BGR2GRAY) # converte para grayscale
#ret2,f = cv2.threshold(f,0,255,cv2.THRESH_BINARY+cv2.THRESH_OTSU)
plt.figure(figsize=(10, 10))
mm.show(a)  
```
