# mmdcalc

```python
img = mm.read('https://drive.google.com/file/d/1c4hnv9hJOV__b99thC-djW3IFps9NXQP/view?usp=sharing')
a = cv2.cvtColor(img ,cv2.COLOR_BGR2GRAY) # converte para grayscale
plt.figure(figsize=(10, 10))
mm.show(a)  
```

```python
b = mm.gradm(a,mm.secross())
plt.figure(figsize=(10, 10))
mm.show(b)  
```

```python
c = mm.openth(a,mm.sebox(3))
plt.figure(figsize=(10, 10))
mm.show(c) 
```

```python
d = mm.threshold(c,140)
plt.figure(figsize=(10, 10))
mm.show(img,d) 
```

```python
e = mm.dil(d,mm.sebox(2))
plt.figure(figsize=(10, 10))
mm.show(img,e) 
```

```python
f = mm.waterB(e,e,mm.sebox(),'line')
plt.figure(figsize=(10, 10))
mm.show(f)
```

```python
f = np.array(f>2,dtype='uint8')
mm.show(img,f)
```

```python
g = np.array(np.maximum(e,f)>0, dtype='uint8')*255
plt.figure(figsize=(10, 10))
mm.show(g) 
```

```python
markers = mm.watershed(img,g,'line')
plt.figure(figsize=(10, 10))
mm.show(markers)
```
