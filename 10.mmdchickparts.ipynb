```python
img = mm.read('https://drive.google.com/file/d/1FCJuM0V-5za3GuKpY5fn9geBKQBbhMln/view?usp=sharing')
a = cv2.cvtColor(img ,cv2.COLOR_BGR2GRAY) # converte para grayscale
plt.figure(figsize=(7, 7))
mm.show(a)
```

```python
b = mm.threshold(a,100)
#b = mm.close(b)
#b = mm.open(b)
b = mm.clohole(b)
plt.figure(figsize=(7, 7))
mm.show(b)
```

```python
plt.figure(figsize=(7, 7))
c = mm.blob(b,'area')
```

```python
plt.figure(figsize=(10, 10))
c0 = mm.blob(b,'textArea')
```

```python
m = mm.blob(b,'area',1,0.01,False)
m
```

```python
wings = np.array(np.logical_and(50<c, c<2500),dtype='uint8')*255
plt.figure(figsize=(7, 7))
mm.show(wings)
```

```python
tights = np.array(np.logical_and(2500<c, c<5500),dtype='uint8')*255
plt.figure(figsize=(7, 7))
mm.show(tights)
```

```python
legs = np.array(np.logical_and(5500<c, c<8500),dtype='uint8')*255
plt.figure(figsize=(7, 7))
mm.show(legs)
```

```python
breast = np.array(8500<c,dtype='uint8')*255
plt.figure(figsize=(7, 7))
mm.show(breast)
```
