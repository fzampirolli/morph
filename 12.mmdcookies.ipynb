# mmdcookies

```python
img = mm.read('https://drive.google.com/file/d/15A50BdGyM7BBs0JUaofG7dPLxi-a7jAL/view?usp=sharing')
a = cv2.cvtColor(img ,cv2.COLOR_BGR2GRAY) # converte para grayscale
plt.figure(figsize=(7, 7))
mm.show(a)
```

```python
b = mm.threshold(a)
plt.figure(figsize=(7, 7))
mm.show(b)
```

```python
aux = mm.blob(b,'circle')
plt.figure(figsize=(7, 7))
mm.show(aux)
```

```python
c = mm.openth(b,mm.sedisk(130))
plt.figure(figsize=(7, 7))
mm.show(c)
```

```python
d = mm.areaopen(c,300)
plt.figure(figsize=(7, 7))
mm.show(d)
np.max(d)
```

```python
plt.figure(figsize=(7, 7))
mm.show(img,mm.gradm(d.astype('uint8')))
```

```python
e = np.logical_xor(b,(d>0).astype('uint8')).astype('uint8')
f = mm.open(e)
plt.figure(figsize=(7, 7))
mm.show(f)
```

```python
plt.figure(figsize=(7, 7))
mm.show(img,mm.gradm(f),mm.gradm(d.astype('uint8')))
```
