```python
img = mm.read('https://drive.google.com/file/d/1mWEp7dmVANTKHGte0ccFzYjxpDn7RN_s/view?usp=sharing') 
a = cv2.cvtColor(img ,cv2.COLOR_BGR2GRAY) # converte para grayscale
plt.figure(figsize=(7, 7))
mm.show(a)
```

```python
b = mm.dil(a,mm.sedisk(5))
d = mm.dil(b,np.ones((65,5),dtype='uint8'))
plt.figure(figsize=(7, 7))
mm.show(d)
```

```python
d = mm.hmin(d,20)
f = mm.neg(mm.regmin(d))
plt.figure(figsize=(7, 7))
mm.show(f)
```

```python
g = mm.watershed(img,f)
plt.figure(figsize=(7, 7))
mm.show(img,mm.dil(np.array(g==-1,dtype='uint8')))
```
