```python
img = mm.read('https://drive.google.com/file/d/1XooqMDovUKR5WTLaBIny3Gmqcm7_W819/view?usp=sharing')
a = cv2.cvtColor(img ,cv2.COLOR_BGR2GRAY) # converte para grayscale
#ret2,f = cv2.threshold(f,0,255,cv2.THRESH_BINARY+cv2.THRESH_OTSU)
plt.figure(figsize=(7, 7))
mm.show(a)
```

```python
B = np.ones((50,90),dtype= 'uint8')
#b = mm.closeth(a,B)
b = cv2.morphologyEx(a, cv2.MORPH_CLOSE, b) - a
plt.figure(figsize=(7, 7))
mm.show(b)
```

```python
c0=mm.closerec(b,mm.sebox(0),mm.secross())
mm.show(c0)
```

```python
c = mm.asf(c0,'CO',mm.secross(0),3)
plt.figure(figsize=(7, 7))
mm.show(c)
```

```python
d = np.array(c>175)
plt.figure(figsize=(7, 7))
mm.show(d)
```
