# mmdholecenter

```python
img = mm.read('https://drive.google.com/file/d/1RRI13mt5GEF7omRo2ngzsddl8aDZ7FpL/view?usp=sharing')
a = cv2.cvtColor(img ,cv2.COLOR_BGR2GRAY) # converte para grayscale
ret2,a = cv2.threshold(a,0,255,cv2.THRESH_BINARY+cv2.THRESH_OTSU)
plt.figure(figsize=(7, 7))
mm.show(a)
```

```python
b = mm.clohole(a)
plt.figure(figsize=(7, 7))
mm.show(b)
```

```python
c = mm.dist(b)
plt.figure(figsize=(7, 7))
mm.show(c>25)
```

```python
e = mm.regmax(c,mm.secross(5))
plt.figure(figsize=(7, 7))
mm.show(e)
```

```python
f = mm.infrec(c>25,b)
plt.figure(figsize=(7, 7))
mm.show(f)
```

```python
g = mm.intersec(a,f)
plt.figure(figsize=(7, 7))
mm.show(g)
```

```python
h = mm.thin(g)
plt.figure(figsize=(7, 7))
mm.show(img,h)
```
