```python
img = mm.read('https://drive.google.com/file/d/1dCJ2UWKLRTscEWo3Qg9ve16Fip64iYYs/view?usp=sharing')
a = cv2.cvtColor(img ,cv2.COLOR_BGR2GRAY) # converte para grayscale
plt.figure(figsize=(7, 7))
mm.show(a)
```

```python
f = a
f1 = mm.toggle(f, mm.ero(f,mm.sedisk(2)), mm.dil(f,mm.sedisk(2)))
plt.figure(figsize=(7, 7))
mm.show(f1)
```

```python
b = np.array(f1>70,dtype='uint8')*255
plt.figure(figsize=(7, 7))
mm.show(b)
```

```python
b = np.array(mm.areaopen(b,10)>0,dtype='uint8')*255
c = mm.close(b)
plt.figure(figsize=(7, 7))
mm.show(c)
```
