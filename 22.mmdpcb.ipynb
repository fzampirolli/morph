# mmdpcb

```python
img = mm.read('https://drive.google.com/file/d/1LtOnintpbbOWwkt1Nca1Hkq2ubcqSiIA/view?usp=sharing')
a = cv2.cvtColor(img ,cv2.COLOR_BGR2GRAY) # converte para grayscale
a = mm.threshold(a)
mm.show(a)
```

```python
b = mm.clohole(a)
holes = mm.subm(b,a)
mm.show(img, holes)
```

```python
c = mm.open(b,mm.sebox(7))  # pode usar tb o edgeoff
c2 = mm.edgeoff(b)
square = mm.cdil(c, a)
mm.show(img, c2, square)
```

```python
f = mm.subm(b, c);
g = mm.open(f, mm.sedisk(17));
circle = mm.cdil(g,a);
mm.show(img, holes, square, circle)
```

```python
B = np.ones((7,20),dtype= 'uint8')
i = mm.subm(f, g)
m = mm.open(i,B)
rect = mm.cdil(m,a)
mm.show(img, holes, square, circle, rect)
```

```python
o = mm.subm(i,m)
p = mm.open(o, mm.sebox(1))
thin = mm.cdil(p,a)
mm.show(img,holes, square, circle, rect, thin)
```

```python
r = mm.subm(o,p)
s = mm.open(r, mm.sebox())
thick = mm.cdil(s,a)
plt.figure(figsize=(7, 7))
mm.show(img,holes, square, circle, rect, thin, thick)
```
