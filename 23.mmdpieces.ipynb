# mmdpieces

```python
img = mm.read('https://drive.google.com/file/d/1m8rx7ky19LczXTys-i-C1JuCL3nt2qM8/view?usp=sharing')
a = cv2.cvtColor(img ,cv2.COLOR_BGR2GRAY) # converte para grayscale
a = mm.threshold(a)
mm.show(a)
```

```python
b = mm.close(a,mm.secross())
mm.show(b)
```

```python
c = mm.thin((b>0).astype('uint8'))
plt.figure(figsize=(7, 7))
mm.show(c)
```

```python
d = mm.dil(c,mm.sebox())
plt.figure(figsize=(7, 7))
mm.show(d)
```

```python
e = mm.clohole(d)
plt.figure(figsize=(7, 7))
mm.show(e)
```

```python
f = mm.subm(e,d)
plt.figure(figsize=(7, 7))
mm.show(f)
```

```python
g = mm.infrec(f,e)
plt.figure(figsize=(7, 7))
mm.show(g)
```
